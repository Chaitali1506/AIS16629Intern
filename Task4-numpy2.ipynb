```python
#Generate Random Number
#NumPy offers the random module to work with random numbers.
#Generate a random integer from 0 to 100:
from numpy import random
x = random.randint(100)
print(x)
```

```python
#Generate Random Float
#The random module's rand() method returns a random float between 0 and 1.
#Generate a random float from 0 to 1:
from numpy import random
x = random.rand()
print(x)
```

```python
#Generate a 1-D array containing 5 random integers from 0 to 100:
from numpy import random
x=random.randint(100, size=(5))
print(x)
```

```python
#Generate Random Number From Array
from numpy import random
x = random.choice([3, 5, 7, 9])
print(x)
```

```python
#Random Distribution
#A random distribution is a set of random numbers that follow a certain probability density function.
```

```python
#Generate a 1-D array containing 100 values, where each value has to be 3, 5, 7 or 9.
#The probability for the value to be 3 is set to be 0.2
#The probability for the value to be 5 is set to be 0.1
#The probability for the value to be 7 is set to be 0.3
#The probability for the value to be 9 is set to be 0.4
from numpy import random
x = random.choice([3, 5, 7, 9], p=[0.2, 0.1, 0.3, 0.4], size=(100))
print(x)
```

```python
#Create your own ufunc for addition:
import numpy as np
def myadd(x, y):
  return x+y
myadd = np.frompyfunc(myadd, 2, 1)
print(myadd([1, 2, 3, 4], [5, 6, 7, 8]))
```

```python
#Use an if statement to check if the function is a ufunc or not:
import numpy as np
if type(np.add) == np.ufunc:
  print('add is ufunc')
else:
  print('add is not ufunc')
```

```python
#Add the values in arr1 to the values in arr2:
import numpy as np
arr1 = np.array([10, 11, 12, 13, 14, 15])
arr2 = np.array([20, 21, 22, 23, 24, 25])
newarr = np.add(arr1, arr2)
print(newarr)
```

```python
#Subtract the values in arr2 from the values in arr1:
import numpy as np
arr1 = np.array([20, 21, 22, 23, 24, 25])
arr2 = np.array([10, 20, 30, 40, 50, 60])
newarr = np.subtract(arr1, arr2)
print(newarr)
```

```python
#Multiply the values in arr1 with the values in arr2:
import numpy as np
arr1 = np.array([10, 20, 30, 40, 50, 60])
arr2 = np.array([20, 21, 22, 23, 24, 25])
newarr = np.multiply(arr1, arr2)
print(newarr)
```

```python
#Divide the values in arr1 with the values in arr2:
import numpy as np
arr1 = np.array([10, 20, 30, 40, 50, 60])
arr2 = np.array([3, 5, 10, 8, 2, 33])
newarr = np.divide(arr1, arr2)
print(newarr)
```

```python
#Raise the valules in arr1 to the power of values in arr2:
import numpy as np
arr1 = np.array([10, 20, 30, 40, 50, 60])
arr2 = np.array([3, 5, 6, 8, 2, 33])
newarr = np.power(arr1, arr2)
print(newarr)
```

```python
#Find the LCM of the following two numbers:
import numpy as np
num1 = 4
num2 = 6
x = np.lcm(num1, num2)
print(x)
```

```python
#Find the LCM of the values of the following array:
import numpy as np
arr = np.array([3, 6, 9])
x = np.lcm.reduce(arr)
print(x)
```

```python
#Find union of the following two set arrays:
import numpy as np
arr1 = np.array([1, 2, 3, 4])
arr2 = np.array([3, 4, 5, 6])
newarr = np.union1d(arr1, arr2)
print(newarr)
```

```python
#Find intersection of the following two set arrays:
import numpy as np
arr1 = np.array([1, 2, 3, 4])
arr2 = np.array([3, 4, 5, 6])
newarr = np.intersect1d(arr1, arr2, assume_unique=True)
print(newarr)
```
